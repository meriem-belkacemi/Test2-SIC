#### Answer the following questions.

1). Make a frequency table of 'yearOfStart' and visualize by year. 
- Sort by the year.
- Draw a line plot from 1997 to 2014. <= Hint: plt.xlim()

```python
years= df['yearOfStart']
t = years.value_counts()
f_table=t.sort_index()

print (f_table.head(10))

```

```python
fig = plt.figure()
ax = plt.axes()
plt.xlim([1997,2014])
ax.plot(f_table.index, f_table.values)
```

2). Now, split the data by the current state of business ('In' or 'Out' of business). Then, visualize the yearly trend of the 'yearOfStart' frequencies. 
- Sort by the year.
- Draw two overlapping line plots from 1997 to 2014.
- Use the 'figure' object.

```python
new=df.groupby(['CurrentState']).groups.items()
type(new)
```

```python
first=df[df['CurrentState']=="In"]['yearOfStart']
second=df[df['CurrentState']=="Out"]['yearOfStart']
t1 = first.value_counts()
t2 = second.value_counts()
f_table1=t1.sort_index()
f_table2=t2.sort_index()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.xlim([1997,2014])
ax.plot(f_table1.index, f_table1.values, c='red')
ax.plot(f_table2.index, f_table2.values, c='blue')
```
